######                                        Beautiful soup



```python
# Step 1: Importing libraries
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://ravindrababuravula.com/courses.php'
data = requests.get(url)
```

```python
#creating Soup Object

soup = BeautifulSoup(data.text, 'html.parser')
soup
```

```python
#to retrieve data
# all a tags with class title:
for i in soup.find_all('a', class_='title'):
    print(i.text)
```

```python
#find the first a and print its text
soup.find('a', class_='title').text
```

```python
for i in soup.find_all('div', class_='description'):
    print(i.text)
```
