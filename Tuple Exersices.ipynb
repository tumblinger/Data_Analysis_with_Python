```python
#Q1 Create myTuple tuple with the follwoing values ("NPower","JDA","Tuesday",30,31)

myTuple = ("NPower","JDA","Tuesday",30,31)
```

```python
#Q2 What is the type of myTuple

type(myTuple)
```

```python
#Q3 What is the length of myTuple

len(myTuple)
```

```python
#Q4 print the values in each index #Use regular indexing

myTuple[0:6]
```

```python
#Q5 print the values in each index #Use negative indexing

print(myTuple[-1])
print(myTuple[-2])
print(myTuple[-3])
print(myTuple[-4])
print(myTuple[-5])
```

```python
#Q6 what is the type of each value

print(type(myTuple[0]))
print(type(myTuple[1]))
print(type(myTuple[2]))
print(type(myTuple[3]))
print(type(myTuple[4]))
```

```python
#Q7 unpack myTuple in the follwoeing variables name,program,dayName,month,day
# print the variables

name = myTuple[0]
program = myTuple[1]
dayName = myTuple[2]
month = myTuple[3]
day = myTuple[4]
print("Name is " + name)
print("Program is " + program)
print("DayName is " + dayName)
print("Month is " + str(month))
print("Day is " + str(day))
```

```python
#Tuples are immutable
#we can always make the testTuple variable reference a new tuple in the memory
#and hold a different information
testTuple=(1,2,3)
print(testTuple)
testTuple=(4,5,6)
print(testTuple)
#But we can't change or edit a value for the existing tuple
testTuple[0]=6 #ERROR 'tuple' object does not support item assignment

```
