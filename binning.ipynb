#### In our dataset, "horsepower" is a real valued variable ranging from 48 to 288, it has 57 unique values. What if we only care about the price difference between cars with high horsepower, medium horsepower, and little horsepower (3 types)? Can we rearrange them into three ‘bins' to simplify analysis?


```python
df.replace('?', np.nan, inplace=True)
df.dropna(inplace=True)
```

```python
df
```

```python
bins=np.linspace(min(df['horsepower']), max(df['horsepower']),4)
bins
```

```python
#We will use the Pandas method 'cut' to segment the 'horsepower' column into 3 bins and determine
#what each value of df['powerhorse'] belongs to:
groups = ['Little', 'Medium', 'High']

#create a new column called 'binned':
df['binned'] = pd.cut(df['horsepower'], bins, labels = groups, include_lowest=True)
df.head(30)
```

### Use a histogram to visualize the distribution of bins we created above.


```python
import matplotlib.pyplot as plt

plt.hist(df['horsepower'],  edgecolor='black')
```

```python
plt.hist(df['binned'],  edgecolor='black', bins=3)
```
